# Chapter 8. 통계적 모멘트 - 비대칭도와 첨도 (Statistical Moments - Skewness and Kurtosis)

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

우리가 variance 를 구할 때 deviation 을 제곱하므로 실제로 얼마나 음수 혹은 양수 쪽으로 치우쳐 있는지 알지 못한다. 따라서 분포의 비대칭과 대칭 문제가 발생한다. 평균의 양쪽으로 분산 mirror image 를 이루면 <i>대칭</i> 이라고 한다. 정규분포가 한 예이다.  정규분포는 평균 $\mu$ 와 표준편차 $\sigma$ 로 다음과 같이 표시된다.
$$ f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $$
그림으로 확인해 보면,

```python
# 평균 = 0 이고 표준편차 = 2 인 정규분포
xs = np.linspace(-6, 6, 300)
normal = stats.norm.pdf(xs)        #stats.norm : normal continuous random variable, pdf : probability density function
plt.plot(xs, normal);
```

비대칭의 분포를 <i>skewed</i> 되었다고 한다. 양의 방향으로 편향된 분포는 mean > median > mode 의 특성을 가지고 음의 방향으로 편향된 분포는 mean < median < mode 의 특성을 보이고 모두 단봉형 (unimodal) 이다.

skewness 의 공식은:
$$ S_K = \frac{n}{(n-1)(n-2)} \frac{\sum_{i=1}^n (X_i - \mu)^3}{\sigma^3} $$

$n$ : 관측값의 갯수, $\mu$ :  산술평균,  $\sigma$ : standard deviation.

```python
# skewed 된 분산 그래프를 그리기 위해 stats.lognorm (lognormal continuous random variable, ppf (Percent point function )) 사용
xs2 = np.linspace(stats.lognorm.ppf(0.01, .7, loc=-.1), stats.lognorm.ppf(0.99, .7, loc=-.1), 150)

# Negatively skewed distribution
lognormal = stats.lognorm.pdf(xs2, .7)
plt.figure(figsize=(8,4))
plt.plot(xs2, lognormal, label='Skew > 0')

# Positively skewed distribution
plt.plot(xs2, lognormal[::-1], label='Skew < 0')
plt.legend();
```

# Kurtosis

모든 정규분포는 평균과 분산에 무관하게 Kurtosis 3 을 가지고 있다. kurtosis > 3 인 폭이 좁은 분포는 peak 가 높고 fat tail 을 가지고 있다. kurtosis < 3 인  분포는 폭이 넓다. 그러나 때때로 정규분포 (kurtosis 3) 를 초과하는 kurtosis 가 쓰이는데 scipy 는 이 것이 default 이다. 폭이 좁은 분포는 평균으로부터 더 큰 jump 가 빈번하고 폭이 넓은 분포는 그 반대이다. 

```python
# 예제 plot
plt.plot(xs,stats.laplace.pdf(xs), label='Leptokurtic')
print('Excess kurtosis of leptokurtic distribution:', (stats.laplace.stats(moments='k')))
plt.plot(xs, normal, label='Mesokurtic (normal)')
print('Excess kurtosis of mesokurtic distribution:', (stats.norm.stats(moments='k')))
plt.plot(xs,stats.cosine.pdf(xs), label='Platykurtic')
print('Excess kurtosis of platykurtic distribution:', (stats.cosine.stats(moments='k')))
plt.legend();
```

# Jarque-Bera 를 이용한 정규성테스트 (Normality Testing)

Jarque-Bera test 는 skewness 와 kurtosis 가 정규분포에 가까운지 여부를 검사 하는 일반 통계 테스트이다. MSFT 의 return 이 정규분포로 부터 왔는지 측정하기 위해 p-vaue 를 구해본다. Jarque Bera test 의 영가설 (null hypothesis) 은 data 가 정규분포로 부터 왔다는 것이다. 따라서, p-value 가 0.05 보다 작으면 비정규 프로세스에 맞지 않는 쪽이 error 일 수 있다. 안전하게 cutoff 를 올려 test 할 수도 있다.
p-value 는 단순히 그렇다 아니다로만 취급해야지 p-value 자체의 값들을 비교하려 하면 안됨을 기억한다. 여기서는 0.05 를 cutoff 로 사용.

```python
from statsmodels.stats.stattools import jarque_bera

N = 1000
M = 1000

pvalues = np.ndarray((N))

for i in range(N):
    # Draw M samples from a normal distribution 
    X = np.random.normal(0, 1, M);
    _, pvalue, _, _ = jarque_bera(X)
    pvalues[i] = pvalue
    
# count number of pvalues below our default 0.05 cutoff
num_significant = len(pvalues[pvalues < 0.05])

print(float(num_significant) / N)
```

```python
_, pvalue, _, _ = jarque_bera(returns)

if pvalue > 0.05:
    print('The returns are likely normal.')
else:
    print('The returns are likely not normal.')
```
