# Chapter16. Violations of Regression Models

regression 을 data 에 맞추기 위해서는 다음의 조건이 만족되어야 good parameter 계산값과 정확한 fit statistics 를 얻을 수 있다.

* 편향되지 않음 (다른 sample 에서도 예측값이 실제 값임)
* 일관됨(많은 sample 이 실제값에 수렴)
* 효율적 (최소 variance)

아래에서 이러한 가정들이 깨어질 수 있는 사례들과 parameter 와 statstics 에 미치는 영향을 조사하겠다. 단일 변수 선형 방정식을 하나의 sample 에 적용하고, 다른 모델에도 같은 생각을 적용해 보겠다. 또한 우리의 모델이 식으로 정확히 기술되었다고 가정한다. 

# Residuals 에 촛점을 맞출 것

모델을 평가할 때 모델 자체 보다는 residual (error, 예측치와 실제값의 차이) 로부터 많은 정보를 얻을 수 있다. 아무리 모델이 복잡해도 residual 은 계산할 수 있고 residual 이 있으면 많은 통계적 test 가 가능하다.

residual 이 정규분포를 따르지 않으면 예측의 정확성에 무언가 잘못되었음을 알아야 한다.

# 비정규 분포된 Residual (Residuals not normally-distributed)

error term 이 정규분포 되지 않으면, 통계 테스트의 유의성이 없게된다. 다행히도, 중심극한정리는 충분히 큰 sample data 는 계수 분포가 정규분포에 가까워진다는 것을 이야기 해 준다. 따라서, large dataset 에는 우리의 분석이 여전히 valid 하다. (비록 error 는 그렇지 않더라도)

* 중심 극한 정리(中心 極限 定理, 영어: central limit theorem, 약자 CLT)는 동일한 확률분포를 가진 독립 확률 변수 n개의 평균의 분포는 n이 적당히 크다면 정규분포에 가까워진다는 정리


## 정규성 테스트

Jarque-Bera test : `statsmodels.stats.stattools.jarque_bera`로 테스트

### 언제나 정규성을 테스트 해야 한다.

```python
# Import all the libraries we'll be using
import numpy as np
import statsmodels.api as sm
from statsmodels import regression, stats
import statsmodels
import matplotlib.pyplot as plt
```

```python
residuals = np.random.normal(0, 1, 100)

_, pvalue, _, _ = statsmodels.stats.stattools.jarque_bera(residuals)         # normal distribution > 0.05 pvalue
print(pvalue)

residuals = np.random.poisson(size = 100)

_, pvalue, _, _ = statsmodels.stats.stattools.jarque_bera(residuals)        # non-normal distribution < 0.05 pvalue
print(pvalue)
```

```python
plt.plot(residuals);    #poisson
```

# Heteroskedasticity (이분산성)

이분산성은 error term 의 분산이 측정치 전체에 걸쳐 일정하지 않은 것이다. 이 것은 측정값이 regression line 에 균등 분포되어 있지 않다는 것이다.


* 등분산성(homoscedasticity)

영어 단어만 보면 생소하고 어렵지만, 한글 단어로 보면 간단하고 명쾌한 의미이다. 등분산성 혹은 동분산성이라고 하는데, 말 그대로 분산이 동일하다는 뜻이다. 좀 더 자세히 설명하면, 모집단에서 종속변수들이 정규분포를 이루는데 각각의 변수들의 정규분포의 표준편차가 동일하다는 것을 의미한다. 선형회귀분석의 기본 가정으로 등분산성이 존재하기 때문에, 회귀분석을 진행하기 전에 등분산성이 성립되는지 꼭 확인하여야 한다.

 

* 이분산성(heteroscedasticity)

앞에서 설명한 등분산성의 반대 개념으로, 각각의 변수들의 정규분포의 표준편차가 동일하지 않은 경우를 의미한다. 주로 횡단면 자료에서 많이 나타난다.

```python
# Artificially create dataset with constant variance around a line : 등분산성 데이터
xs = np.arange(100)                     # [ 0,  1,  2,  3,  .... 97, 98, 99]
y1 = xs + 3*np.random.randn(100)

# Get results of linear regression
slr1 = regression.linear_model.OLS(y1, sm.add_constant(xs)).fit()

# Construct the fit line
fit1 = slr1.params[0] + slr1.params[1]*xs

# Plot data and regression line
plt.scatter(xs, y1)
plt.plot(xs, fit1)
plt.title('Homoskedastic errors');
plt.legend(['Predicted', 'Observed'])
plt.xlabel('X')
plt.ylabel('Y');
```

위 등분산성 dataset 의 그래프는 regression line 을 따라 error term 이 균일하게 분포되어 있고 x 값의 범위에 영향 받지 않는다.

```python
# Artificially create dataset with changing variance around a line : 이분산성 데이터
y2 = xs*(1 + .5*np.random.randn(100))

# Perform linear regression
slr2 = regression.linear_model.OLS(y2, sm.add_constant(xs)).fit()
fit2 = slr2.params[0] + slr2.params[1]*xs

# Plot data and regression line
plt.scatter(xs, y2)
plt.plot(xs, fit2)
plt.title('Heteroskedastic errors')
plt.legend(['Predicted', 'Observed'])
plt.xlabel('X')
plt.ylabel('Y')

# Print summary of regression results
slr2.summary()
```

이 이분산성 dataset 의 그래프는 error term 이 regression line 에서 불규칙하게 퍼져 있고 x 의 range 에 따라 다르게 분포되어 있다.

### 이분산성 테스트 (Testing for Heteroskedasticity)

이분산성 테스트 방법은 여러가지가 있는데 그 중 Breush Pagan test 가 statsmodels 에서 제공되는데, 정규성 검사도 함께 수행된다. 그러나 이분산성을 가지면서도 정규분포된 residual 을 가질 수 있으므로 test 가 반드시 수행되어야 한다.

statsmodels.stats.stattools.jarque_bera returns :

        Jarque-Bera test statistic, pvalue of the test statistic (> 0.05 이면 normal distibution), skewness, kurtosis 

statsmodels.stats.diagnostic.het_breuschpagan returns:

       lagrange multiplier statistic, p-value of lagrange multiplier test, f-statistic, p-value for the f-statistic
       
       hypothesis 는 residual 이 x 변수에 종속되지 않음 이다. 즉, 제멋대로 분포함. 

```python
residuals1 = y1 - fit1           # y1 = xs + 3*np.random.randn(100), fit1 = slr1.params[0] + slr1.params[1]*xs : 등분산 dataset
residuals2 = y2 - fit2           # y2 = xs*(1 + .5*np.random.randn(100)), fit2 = slr2.params[0] + slr2.params[1]*xs : 이분산 dataset

xs_with_constant = sm.add_constant(xs)         # [[ 1.,  0.], [ 1.,  1.],.... [ 1., 98.],[ 1., 99.]]

_, jb_pvalue1, _, _ = statsmodels.stats.stattools.jarque_bera(residuals1)          # residual 의 정규성 테스트
_, jb_pvalue2, _, _ = statsmodels.stats.stattools.jarque_bera(residuals2)
print("p-value for residuals1 being normal", jb_pvalue1)                                  # pvalue > 0.05 이면 가설(정규분포)이 맞음
print("p-value for residuals2 being normal", jb_pvalue2)                                  # pvalue < 0.05 이면 가설(정규분포) reject

_, pvalue1, _, _ = stats.diagnostic.het_breuschpagan(residuals1, xs_with_constant)  # 등분산 dataset 의 pvalue > 0.05 이면 가설 reject
_, pvalue2, _, _ = stats.diagnostic.het_breuschpagan(residuals2, xs_with_constant)  # 이분산 dataset 의 pvalue < 0.05 이면 가설이 맞음
print("p-value for residuals1 being heteroskedastic", pvalue1)
print("p-value for residuals2 being heteroskedastic", pvalue2)
```

### 이분산성의 교정 (Correcting for Heteroskedasticity)

조건부 이분산성 (위에서와 같이 독립변수에 error variance 가 correlate 되어 있을 경우. (예) x 의 증가에 따라 error variance 증가) 은 regression 유의성에 관한 F-test 와 개별 계수의 유의성을 신뢰할 수 없게 만든다. 가장 흔하게는 fit 의 유의성을 과대평가하게 만든다.

Breusch-Pagan test 와 White test 로 조건부 이분산성 (conditional heteroskedasticity)을 detect 할 수 있다. 이러한 효과가 있는 것으로 의심되면 모델을 수정할 수 있다. 한 방법은 generalized least squares 인데 original equation 을 manually 변경해야 한다. robust standard error 를 계산하는 또다른 방법은 이분산성을 감안한 fit statistic 을 수정하는 것이다. `statsmodels` 은 robust standard errors 를 계산한다 아래에서 그 차이를 보자.

statsmodels.regression.linear_model.OLSResults.get_robustcov_results: create new results instance with robust covariance as default

```python
print(slr2.summary())                                        # 이분산성 dataset 의 linear regression pvalue < 0.05 이면  dataset 이 linear 함
print(slr2.get_robustcov_results().summary())     # robust result 의 pvalue < 0.05 이면  dataset 이 linear 함
```

# Serial correlation of errors ( error 의 자기 상관)

관측치들의 error 가 연관되어 있는 것은 (serial correlation 혹은 자기상관이라고 함)일반적으로 발생하는 중요한 문제이다. 이것은 정기적인 fluctuation 이 있는 time series data 를 이용할 때 자주 마주치는 문제이다. 하나의 독립변수가 종속변수의 이전 값에 종속된다면 - 이전 기간의 종속변수의 값과 같은 경우와 같이 - 또는 부정확한 model 기술이 자기상관을 초래한 경우, 계수는 일관성 없을 것이고 invalid 하게 된다. 이 것은 model 들이 autocorrelation  없다는 가정하에 만들어 졌기 때문이므로, pvalue 등이 invalid 하다고 나타난다.

error 들이 등분산성이라면,  Durbin-Watson test 를 이용하여 자기상관을 테스트할 수 있다.

```python
# Load pricing data for an asset
from googlefinance.client import get_price_data, get_prices_data
import pandas as pd

param_STOCK = {'q': "DAL", 'i': "86400", 'x': "NYSE", 'p': "20Y" }
df = pd.DataFrame(get_price_data(param_STOCK))
df.index = df.index.date
df.index = pd.to_datetime(df.index)

start = pd.to_datetime('2014-01-01')
end = pd.to_datetime('2015-01-01')
y = df['Close'].loc[start : end]

x = np.arange(len(y))

# Regress pricing data against time
model = regression.linear_model.OLS(y, sm.add_constant(x)).fit()

# Construct the fit line
prediction = model.params[0] + model.params[1]*x

# Plot pricing data and regression line
plt.plot(x,y)
plt.plot(x, prediction, color='r')
plt.legend(['DAL Price', 'Regression Line'])
plt.xlabel('Time')
plt.ylabel('Price')

# Print summary of regression results
model.summary()
```

위의 그래프와 같이 spike 가 있는 데이터는 autocorrelate 관계가 분명히 있다. 

### 자기상관 테스트 (Testing for Autocorrelation)

price 와 residual 두가지에 대하여 모두 자기상관을 테스트해 보자. Ljun-Box test 는 n 번째 lagged datapoint 가 현재를 예측하는지의 확률을 계산한다. max lag 가 주어지지 않으면 최대 lag 를 계산하고 그 lag 까지의 모든 lags 에 대한 p-value 를 반환한다. 여기서는 5 개의 최근 datapoint 에 현재가에 대한 유의한 상관관계가 있음을 볼 수 있다. 그러므로 price 와 residual 모두 자기 상관관계가 있다고 결론낼 수 있다.
재미삼아 정규성 테스트도 해 본다.

```python
_, prices_qstats, prices_qstat_pvalues = statsmodels.tsa.stattools.acf(y, qstat=True)
_, prices_qstats, prices_qstat_pvalues = statsmodels.tsa.stattools.acf(y-prediction, qstat=True)

print('Prices autocorrelation p-values', prices_qstat_pvalues)
print('Residuals autocorrelation p-values', prices_qstat_pvalues)

_, jb_pvalue, _, _ = statsmodels.stats.stattools.jarque_bera(y-prediction)

print('Jarque-Bera p-value that residuals are normally distributed', jb_pvalue)
```

## Newey-West

Newey-West 는 자기상관에 대한 variance 를 계산하는 방법이다. 자기상관이 있는 경우, 단순한 계산은 부정확한 standard error 를 만들어 낸다. 

자기상관을 제거하기 위해 regression 식을 바꾸는 시도도 가능하다. 간단히 고치는 방법은 적절한 방식으로 standard error 를 교정하여 유의성 check 를 위해 교정된 값을 사용하는 는 것이다.
아래에서는 statsmodel 의 Newey-West method 를 이용하여 계수의 standard error 를 교정한다. 원래 regression 에서 report 된 것보다 high 한데, 우리가 양의 correlated error 를 예측한 대로이다. 

```python
from math import sqrt

# Find the covariance matrix of the coefficients
cov_mat = stats.sandwich_covariance.cov_hac(model)

# Print the standard errors of each coefficient from the original model and from the adjustment
print 'Old standard errors:', model.bse[0], model.bse[1]
print 'Adjusted standard errors:', sqrt(cov_mat[0,0]), sqrt(cov_mat[1,1])
```

# 예: Anscombe's quartet

Anscombe 은 동일한 평균과 variance 를 가진 4 개의 datset 를 만들었는데, 같은 상관계수와 regression line 및 R-squared error 값을 가지고 있다.
아래에 4 개 dataset 을 plot 하여 비교했는데, 한눈에 보아도 첫번째 dataset 만이 regression model 을 만족시키는 것을 알수 있다. 결론적으로 다른 3개의 high R-squared error 는 의미가 없고, regression line 에 fit 하지 않음을 알 수 있다.

```python
from scipy.stats import pearsonr

# Construct Anscombe's arrays
x1 = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
x2 = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
x3 = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]

# Perform linear regressions on the datasets
slr1 = regression.linear_model.OLS(y1, sm.add_constant(x1)).fit()
slr2 = regression.linear_model.OLS(y2, sm.add_constant(x2)).fit()
slr3 = regression.linear_model.OLS(y3, sm.add_constant(x3)).fit()
slr4 = regression.linear_model.OLS(y4, sm.add_constant(x4)).fit()

# Print regression coefficients, Pearson r, and R-squared for the 4 datasets
print('Cofficients:', slr1.params, slr2.params, slr3.params, slr4.params)
print('Pearson r:', pearsonr(x1, y1)[0], pearsonr(x2, y2)[0], pearsonr(x3, y3)[0], pearsonr(x4, y4)[0])
print('R-squared:', slr1.rsquared, slr2.rsquared, slr3.rsquared, slr4.rsquared)

# Plot the 4 datasets with their regression lines
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)
xs = np.arange(20)
ax1.plot(slr1.params[0] + slr1.params[1]*xs, 'r')
ax1.scatter(x1, y1)
ax1.set_xlabel('x1')
ax1.set_ylabel('y1')
ax2.plot(slr2.params[0] + slr2.params[1]*xs, 'r')
ax2.scatter(x2, y2)
ax2.set_xlabel('x2')
ax2.set_ylabel('y2')
ax3.plot(slr3.params[0] + slr3.params[1]*xs, 'r')
ax3.scatter(x3, y3)
ax3.set_xlabel('x3')
ax3.set_ylabel('y3')
ax4.plot(slr4.params[0] + slr4.params[1]*xs, 'r')
ax4.scatter(x4,y4)
ax4.set_xlabel('x4')
ax4.set_ylabel('y4');
```
