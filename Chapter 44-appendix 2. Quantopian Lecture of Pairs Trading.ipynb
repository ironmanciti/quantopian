## Introduction

Pairs trading 은 highly correlated 되어 있는 instruments (ex. two stocks, exchange-traded funds (ETFs), currencies, commodities or options) 를 쌍으로 구성하여 long position 과 short position 을 match 시키는 market-neutral trading strategy 이다.  Pairs traders 는 correlation  의 weakness 를 기다려 under-performer 를 long 하고, 동시에 over-performer 를 short selling 한 후, relationship 이 statistical norms 로 return 하면 position 을 close 한다.

수익은 두 자산의 각가의 움직이는 방향이 아니라 price change 차이에서 창출된다. 따라서, long position 이 short 보다 더 위로 오르거나, short position 이 long 보다 더 내려가면 profit 이 실현된다. (완전한 상황은 long position 은 오르고 short position 은 떨어지는 것이지만 반드시 그래야 수익이 나는 것은 아니다.) Paris trader 들은 여러가지 시장 조건(강세시장, 약세시장, 횡보 국면) 에서 수익을 낼 수 있고 low or high volatility 기간에도 수익을 낼 수 있다.

Paris Trading 의 기원.

Paris Trading 은 1980 년 중반 모건스탠리의 computer scientists, mathematicians and physicist 가 모여서 만들었다. 연구의 핵심은 역사적으로 유사한 가격 움직임을 보이거나 highly correlated  된 주식의 쌍들을 구분하는 quantitative methods 를 개발하는 것 이었다. 그 팀의 결과물인 black box 는 1987 년에 성공적으로 작동하여 5천만불의 수익을 냈으나 이듬해에 형편없는 결과가 나와 1989 년에 해체되었다.

수년에 걸쳐, paris trading 은 market-neutral investment strategy 로서 individual, institutional and hedge fund traders 의 주목을 끌었는데, 그 이유는 internet 의 출현과 trading 기술의 발달에 힘입은 것이다. 이 두가지 요인은 individual investors 들이  real-time market data 와 powerful tool 을 institutional traders 보다 저렴하게 이용할 수 있게 해 주었다. 물론, large hedge funds 와 기관 trader 들이 여전히 유리하다. (예를 들어 더 강고한 system 과 규모의 경제) 그러나, 오늘날의 시장 참여자들(retail trader 혹은 quant shop 의 highly-skilled mathematicians) 은 real-time market data 접근, trading system 직접 접근, 앞선 computer modeling 과 복잡한 trading strategy 의 자동화가 가능하다.

기술을 이용하여 - fundamental 분석, 확률, 통계 및 기술적 분석 - pairs trader 들은 두개 instruments 의 relationship 들 구분하려 시도하고, relationship 의 방향을 정하고, 제시된 data 에 근거하여 trading 을 수행한다.

## Market Neutral Investing

Paris Trading 은 market-neutral strategy 이다. “Market-neutral investing“ 이란 instrument 의  actual 과 이론적 relationshop 에 따라 long and short position 을 offsetting 하여 market risk 의 neutralize 를 추구하는 것이다. 이러한 접근은 macroeconomic variable 과 market sentiment 의 shift 에 의해 생기는 price 의 시스템적 변화에 대한 exposure 를 제한하는 것을 추구한다.

Market-neutral investing 은 단일한 전략만 있는 것이 아니고 다음과 같이 많은 market-neutral strategies 가 있다.

Convertible arbitrage

Equity hedge

Equity market neutral

Fixed-income arbitrage

Merger arbitrage

Mortgage-backed securities arbitrage

Relative value arbitrage

Statistical arbitrage (“StatArb”)

asset type 에 따라 여러가지 다른 market-neutral strategy 를 사용한다.  

Market-neutrality 는 개별 instrument 혹은 portfolio level 에서 얻어질 수 있다. 

### How market-neutral relates to pairs trading

One position 이 another position 과 연관되어 directional exposure 를 감소시키므로, market-neutral strategy 는 종종 market risk 에 대한 hedge 를 제공한다. 이와 같은 방식으로 , market 에 대한 exposure 는 long and short position 사이의 관계에 대한 exposure 로 교환된다. 이 말은 market-neutral investing 이 risk-neutral 하거나 risk-free 하다는 뜻은 아니다. (둘다 아님). 그러나, market-neutral strategy 의 risk 는 directional, long-only investing 에 관련된 risk 와 다르다. market-neutral approach 는 overall investment strategy 의 일부로 사용될 경우 alternative and uncorrelated source of returns 을 제공한다.

Paris trader 들은 특정 sector 혹은 industry 의 one stock 을 long 하고 same sector 혹은 industry 의 equal-dollar-value (or dollar neutral) 의 correlated stock 의 short position 에 pairing 하여 방향성 risk 를 제한한다. 

## Correlation

Correlation 은 dependent variable 과 independent variable 간의 relationship 의 강도를 기술하기 위해 linear regression analysis 에서 사용하는 용어이다. Pairs trading 의 중심 idea 는 두 주식 (혹은 instrument) 이 충분히 correlated 되어 있고, correlation 의 변화는 pair’s mean trend 로 reversion 하여 profit 기회를 제공한다는 것 이다. 예를 들어, stock A 와 stock B 가 highly correlated 되어 있고 일시적으로 correlation 이 약화되어 stock A 가 상승하고 stock B 가 하락하면 pair trader 는 이 divergence 를 이용하여 stock A (over-performer) 를 shorting 하고 stock B (under-performer) 를 longing 한다. stock 이 statistical mean 으로 revert 하면 trader 는 profit 을 얻을 수 있다.

###  importance of correlation

Correlation 은 두 개 instrument 의 relationship 을 측정한다. 아래의 daily chart 는 S&P 500 선물과 Dow 선물 계약이 함께 움직이는(correlated 된) 경향을 보여준다.




pairs traders 는 다음을 attempt 한다.

* 두 instruments 간의 relationship 을 identify.
* relationship 의 방향을 determine.
* 제시된 data 를 근거로 execute trades.

rates of return 이나 historical prices 와 같은 두 variables 간의 correlation 은 이들 variable 들이 함께 움직이는 정도에 대한 relative statistical measure 이다. correlation coefficient 는 one variable 의 value 가 다른 것의 value 와 연관된 정도를 측정한다. correlation coefficient 의 범위는 -1 에서 +1 사이이다. -1 은 완전히 반대 움직임, +1 은 완전히 같이 움직이는 것이다. 0 는 no correlation.

Pairs traders 는 price 가 같이 움직이는 경향이 있는, 즉 prices 가 correlated 된 instruments 를 찾는다. 임의의 두 주식의 완전한 positive correlation 을 지속적으로 유지하는 것은 사실 어려운 일이다 (또한 거의 불가능하다) 왜냐하면, 가격들이 완전히 서로를 흉내내야 하기 때문이다. 대신 pairs trader 는 high degree of correlation 을 가진 주식들을 찾아 가격들이 statistical norm 의 outside behavior 를 보일 때 profit 을 attempt 한다. 0.8 correlation 은 pairs traders 의 benchmark 로 종종 사용된다. 0.5 이하이면 일반적으로 weak 한 correlation 으로 간주한다.

이상적으로, good correlation 은 multiple time frames 를 거쳐 나타내진다.

왜 correlation 이 pairs trading 에서 중요한가 ? 두 instruments 가 correlated 되어있지 않으면 가격의 divergence 나 이어지는 convergence 가 일반적으로 덜 meaningful 하다.

 
### Determining correlation

suitable pairs 를 찾아내는 첫 단계는 어떤 공통점이 있으며 good liquidity 가 있으며 short 가 가능한 주식을 찾는 것이다. market risk 가 비슷하므로, 동일한 sector market 내의 경쟁 회사들은 natural potential paris 를 이루고 good place to start 가 된다. 예를 들어 다음과 같은 회사들이다.

Dell and Hewlett-Packard

Duke Energy and NextEra Energy

E-mini S&P 500 and E-mini Dow

Exxon and Chevron

Lowe’s and Home Depot

McDonald’s and Yum! Brands

다음 단계는 그 것들이 어떻게 correlated 되어 있는지 결정하는 것이다. 두 주식이 얼마나 상호 연관되었는지 반영하는 correlation coefficient 를 이용하여 측정할 수 있다. Excel 의 CORREL 함수를 이용하여 계산할 수 있다.



Correlation coefficient 가 정해지고나면, 그 결과를 filter 로 이용하여 가장 potential 있는 pair 를 찾아낸다.
 
### Price ratio

Correlated pairs 를 찾아내고 나면, 관계가 mean reverting 인지 결정할 수 있다. 즉, price 가 diverge 하면 statistical norm 으로 revert 하는가 ? pair's price ratio 를 plotting 하여 정할 수 있다. Correlation coefficient 와 마찬가지로 
 
Price ratio = Price of Instrument A / Price of Instrument B
 

 
 Standard deviation line 을 더하면, price 가 mean 으로부터 얼마나 떨어져 움직이는가 insight 를 얻을 수 있다. 

68.26 percent of the data will fall within +/- one standard deviation of the mean.

95.44 percent of the data will fall within +/- two standard deviations of the mean.

99.74 percent of the data will fall within +/- three standard deviations of the mean.

이 data 를 적용하여, price ratio 가 std 의 x 배 만큼 diverge 될 때까지 기다렸다가 (예를 들어 +/- 2 std) long/short trade 를 들어간다. (std number 의 선택은 historical analysis 와 optimization 으로 결정한다) Pair 가 mean trend 로 revert 하면 trade 는 profitable 할 수 있다.

### Events that trigger weakness in correlation

두개의 instruments 가 highly correlated 되어 있을 때, 어떤 event 가 일시적으로 correlation 을 약화시킬 수 있다. (예를 들어 연방은행의 발표 혹은 지정학적 혼란 등). correlation 의 weakness 를 trigger 하는 events 는 일반적으로 하나의 instrument 에만 제한된 primary impact 을 주게 된다. 예를 들어 한 주식 내의 수요, 공급의 일시적 변화를 들 수 있다.

Changes related to the company’s financial health.

Restructuring or mergers.

Significant information about its products (whether positive or negative).

Changes in key management.

Legal or regulatory issues that could affect the company’s power to conduct business.

## Arbitrage and Pairs Trading

기본적으로는, arbitrage 는 가격 차이의 이점을 살리기 위해 같은 (혹은 동등한) 주식을 다른 시장에서 동시 buying 및 selling 하여 이익을 내는 process 이다. 가격 차이는 market inefficiency 의 결과, pricing mismatches 및 currency exchange rate 의 결과일 수 있다.

Arbitrageur 는 예를 들어 NYSE 에서 ABC 주식을 \$50 에 사고 동시에 LSE (London Stock Exchange) 에서 \$51 에 팔아 주당 1\$의 이익을 낸다. Arbitrageur 는 매우 작은 price 움직임으로부터 이익을 내므로 substantial profit 을 위해 전형적으로 large position 을 가져간다.

Technology 가 발전하기 이전에는, 이러한 차익거래 기회가 많았으나 오늘날과 같이 경기장이 평평해지고, 더 많은 사람들이 market data 에 realtime access 를 하게되어 투명성이 증가함에 따라 많은 순수 차익거래 기회가 더 이상 존재 않게 되었다.
 
순수 차익거래가 risk-free 이고 실제 pricing 결함에 기초한 반면, 다른 형태의 차익거래는 속성상 투기적이고 percieved 혹은 implied pricing flaw 에 기초한다 - historical average 에 비해 상당히 price relationship 이 deviate 되었다는 investor 의 perception. Pairs trading 은 다음 두 가지 arbitrage type 의 특징을 share 한다 - 상대 가치 arbitrage 와 통계적 arbitrage.

* Relative value arbitrage

상대 가치 차익거래는 profit 이 instruments 의 가격간의 favorable change 에 의존할 때, related instruments 를 동시에 사거나 파는 것을 말한다. 이러한 type 의 차익거래는역사적으로 혹은 수학적으로 interrelated 된 주식들이 일시적으로 misalign 되었을 때 position 을 offset 하게 된다. 투자자는 주식간의 관계가 norm 으로 revert 될 때 수익을 실현하게 된다.
 
Relative value arbitrage approaches 는 다음과 같은 여러가지 다른 투자 전략을 포괄한다.

Capital structure arbitrage

Convertible arbitrage

Equity statistical arbitrage

Fixed-income arbitrage

Merger arbitrage

Options and warrants

Pairs trading

Statistical arbitrage

Pairs trading 에는 relative value 와 statistical arbitrage 두가지 요소가 모두 포함되어 있고, 실제로 그 두가지로 불리운다.  

## Fundamental and Technical Analysis for Pairs Trading

Pairs trader 는 어떠한 instrument 를 pair 할 것인가 결정하기 위해 또한 언제 trade 를 in and out 할 것인지 결정하기 이해 fundamental 혹은 technical analysis (또는 둘 다) 를 이용한다. 많은 paris trader 들은 기술적 분석 기술을 적용하여 pair 를 찾아내고 fundamental 을 이용하여 확인한다. 이 여분의 “layer” of analysis 는 trade 가 “makes sense.” 하다는 것을 확신하기 위해 사용될 수 있다. 예를 들어, 모든 기술적 분석이 stock ABC 를 long 하고 XYZ 을 short 하라고 지시하지만 fundamental 이 stock ABC 가 weak earnings 를 발표할 것이라는 것을 보여주면 position 을 재고해야 한다.

### Fundamental factors

Fundamental analysis 는 주식의 value  및 어떤 주식이 단기적으로 better performing 할지 평가하는데 사용되는 관련된 경제적, 재무적 및 기타 양적, 질적 factor 를 말한다. Fundamental analyst 들은 pairs trading 기회를 identify 할 때 다음 factor 들을 고려한다:

Changes in operating margins

Discounted cash flow

Dividend discount model

Dividend yield

Excess cash flow

Float

Price/earnings to growth (PEG ratio)

Price-earnings ratio (P/E ratio)

Price-to-book ratio (P/B ratio)

Price-to-cash-flow ratio (price/cash)

Price-to-sales ratio (price/sales)

Return on equity

Total assets over sales (See also: 5 Basic Financial Ratios and What They Reveal.)

## Pairs Trading: Risks

순수한 차익거래가 essentially risk-free 전략이지만, pairs trading 에도 model risk 와 execution risk 를 포함한 risk 가 있다.

* Model risk

거의 모든 투자에 risk 가 있듯이 pairs trading 은 model risk 에 expose 되어 있다. model risk 는 전략을 생성하는 model 이 기대한대로 수행되지 않을 때 발생한다. 부정확한 research 로 부터 흠결있는 logic 이나 계산에 이르기 까지 여러가지 이유가 있다. 대표적인 예가 Long Term Capital Management (LTCM) 에서 발생한 model risk 이다.

* Execution risk

이 type 의 risk 는 pairs trading 의 return 에 negative impact 을 주는 또 다른 factor 이다. Execution risk 는 전략이 계획대로 실해되지 않을 가능성을 말한다. 예를 들어 trader 는 가격상 slippage 혹은 order 의 부분 체결을 경험할 수 있고 profit potential 이 줄어든다. 

## Disadvantages of Pairs Trading

Pairs trading 과 관련된 risk 외에도 trader 들이 알아야 할 몇가지 disadvantage 들이 있다. 아마도, 가장 뚜렷한 disadvange 는 commission 과 fee 를 이중으로 내야하는 것일 것이다. Pairs trader 는 각 trade 에 enter, exit 할 때 두 개의 commission 을 지불해야 한다. trade 의 전략과 빈도에 따라 빠르게 증가할 수 있다. 전략이 실제 profit 을 낼 수 있는지 결정하기 위해 historical modeling 에 commission 을 반영해야 한다.

* Execution Risk 
 
Execution risk 역시 pairs trading 의 disadvantage 이다. Slippage 나 partial filling 및 bid-ask spread 는 profit 을 감소시킬 수 있다. 

많은 pairs trading 전략이 매우 작은 금액 변화를 긇어모으는 것이므로, large position 을 enter 할 수 있는 자본력이 큰 trader 에게 유리하다. (leverage position) Small traders 들은 large, institutional pairs trader 들과 다른 approach 를 해야 한다.

## Advantages of Pairs Trading

Pairs trading 은 controlled risk, market direction 과 상관없는 profit, smaller drawdown 등 여러 가지 유리한 점이 있는 market-neutral strategy 이다. 

* Controlled risk

Pairs trading 의 중심은 long position 과 related 또는 correlated instruments 의 short position 의 matching 이다. 이 것은 자동으로 한쪽 trade 가 다른 쪽의 hedge 가 되도록 하고 risk 가 꽤 control 된다. Stock ABC 를 long 하고 stock XYZ 을 short 하는 pairs trade 를 예로 들면, 두 개의 stock 이 correlated 되어 있으므로, 전체 sector  가 충격을 받으면 두 stock 의 주가 모두 떨어진다. long 한 주식에서 발생한 손실은 short position 에서 mitigate 된다 따라서, 특정일에 전체 sector 가 down 되더라도, market 평균에 대한 low correlation 으로 말미암아 trader 의 net position 은 neutral 로 남게 된다.

* Profit regardless of market direction

Pairs trading 의 또 다른 feature 는 market 이 오르던, 내리던 또는 횡보하던 profit 을 만드는 능력이다. 이 것은 전략이 market direction 에 종속되지 않고, 두 instrument 의 관계에 종속되기 때문이다. 바꾸어 말하면, long-only trader (bullish) 는 rising market 에서만 이익을 낼 수 있고, Short-only trader(bearish) 는 market 이 하락할 때만 이익을 낼 수 있지만, pairs trader 는 market 의 direction 에 무관하게 두 instrument 의 상대적 performance 에서 결과를 얻는다.

* No directional risk

방향성 risk 는 가격 움직임의 direction 에 위험 노출되는 것이다. 예를 들어, long position 은 price 의 move down 에 risk 노출이 된다. Pairs trading 에서는 2nd instrument 가 1st 에 대해 hedge 로 작용 하므로 directional risk 를 제거한다. 

* Smaller drawdowns

drawdown 은 open 된 investment value 의 peak-to-trough decline 을 말한다. 예를 들어, trade 가 특정일에 \$10,000 의 미실현 이익에 이르고 다음 날 \$6,000 으로 떨어졌다면, drawdwon 은 \$4,000 이다. drawdown 이 trade 가 궁극적으로 winner 가 될 것인지 말해주는 것은 아니지만, 잃을 위험성 있는 돈이므로 drawdown 의 관리는 중요하다. 더우기,  전략의 최대 기대 drawdown (historical model 에 근거한) 은 trade 가 특정 전략에 얼마나 많은 금액을 allocate 해야 할지 결정한다. 예를 들어, 전략의 최대 drawdown 이 \$5,000 이라면, 어떠한 잠재 손실을 cover 하기 위해 최소 \$5,000 이 있어야 한다. losing position 이 winning position 으로 완화되므로 pairs trading 은 smaller net drawdown 을 가지게 된다.

## Pairs Trading: Conclusion

Pairs trading 은 두개의 realted instruments 의 price change 차이에서 profit 을 추구하는 market-neutral investment strategy 이다. 만약 두개의 instrument 가 함께 움직이는 historical tendency 를 가지고 있다면, 어떤 기간의 diverge 후에는 다시 move together 할 것이다. price 가 re-converge 할 때가 trader 가 이익을 내는 시점이다. 

* Pairs trading: Right for your portfolio?
